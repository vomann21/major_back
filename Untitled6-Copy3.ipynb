```python
def remove_special_characters(text):
    import re
    # Replace all non-alphabetic characters and spaces with an empty string
    return re.sub(r'[^a-zA-Z0-9\s]', '', text)
new_df['price'] = new_df['price'].apply(remove_special_characters)
new_df['description'] = new_df['description'].apply(remove_special_characters)
new_df['brand'] = new_df['brand'].apply(remove_special_characters)
new_df['platform'] = new_df['platform'].apply(remove_special_characters)
new_df['type'] = new_df['type'].apply(remove_special_characters)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
# from sklearn.feature_extraction.text import CountVectorizer
# cv = CountVectorizer(max_features=5000,stop_words='english')         ##word count

# ###SKIP GRAM MODEL
# from gensim.models import Word2Vec
# from sklearn.feature_extraction.text import CountVectorizer

# # Assuming new_df['tags'] contains your text data
# cv = CountVectorizer(max_features=5000, stop_words='english')
# vectors = cv.fit_transform(new_df['tags']).toarray()
# word_list = cv.get_feature_names_out()

# # Convert vectors into list of sentences where each sentence represents a document
# sentences = [[word_list[idx] for idx, val in enumerate(row) if val > 0] for row in vectors]

# # Train the Word2Vec model
# model = Word2Vec(sentences, sg=1)  # sg=1 for skip-gram, sg=0 for CBOW

# # Now you can use the trained Word2Vec model to get word embeddings
# word_embeddings = model.wv
# print(word_embeddings)

# # TF-IDF
# from sklearn.feature_extraction.text import TfidfVectorizer
# # Assuming df is your DataFrame and 'tag' is the column with the text
# vectorizer = TfidfVectorizer(lowercase=True, token_pattern=r'\b\w\w+\b', stop_words='english', min_df=1, norm='l2')
# X = vectorizer.fit_transform(new_df['tags']).toarray()


#BERT
# from sentence_transformers import SentenceTransformer

# # Initialize the BERT model (you can replace 'bert-base-nli-mean-tokens' with any other BERT model)
# model = SentenceTransformer('bert-base-nli-mean-tokens')

# # Apply the model to your data
# embeddings = model.encode(new_df['tags'], convert_to_tensor=True).toarray()



#Skip gram model
from gensim.models import Word2Vec
tokenized_text = new_df['tags'].apply(lambda x: x.split())

# Train Skip-Gram model
skipgram_model = Word2Vec(sentences=tokenized_text, vector_size=100, window=5, sg=1, min_count=1)

# Generate Product Embeddings
def get_product_embedding(text, model):
    # Average the word embeddings in the text to get the text embedding
    word_embeddings = [model.wv[word] for word in text if word in model.wv]
    if word_embeddings:
        return np.mean(word_embeddings, axis=0)
    else:
        return np.zeros(model.vector_size)  # Return zero vector if no embeddings found

new_df['embedding'] = tokenized_text.apply(lambda x: get_product_embedding(x, skipgram_model))

# Calculate Similarity
from sklearn.metrics.pairwise import cosine_similarity

# Convert embeddings to numpy array
embeddings = np.stack(new_df['embedding'])
```
