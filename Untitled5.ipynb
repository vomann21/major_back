```python
def remove_special_characters(text):
    import re
    # Replace all non-alphabetic characters and spaces with an empty string
    return re.sub(r'[^a-zA-Z0-9\s]', '', text)
```

```python
df_shuffled['price'] = df_shuffled['price'].apply(remove_special_characters)
df_shuffled['description'] = df_shuffled['description'].apply(remove_special_characters)
df_shuffled['brand'] = df_shuffled['brand'].apply(remove_special_characters)
df_shuffled['platform'] = df_shuffled['platform'].apply(remove_special_characters)
df_shuffled['type'] = df_shuffled['type'].apply(remove_special_characters)
```

```python
def remove_stopwords(tokens):
    # Get English stopwords
    stop_words = set(stopwords.words('english'))
    # Remove stopwords from the tokens
    filtered_tokens = [token for token in tokens if token.lower() not in stop_words]
    return filtered_tokens
```

```python
filtered_tags_tokens = [remove_stopwords(tokens) for tokens in tag_tokens]
```

```python
filtered_tags_tokens
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
def lemmatize_tokens(tokens):
    # Lemmatize each token
    lemmatized_tokens = [lemmatizer.lemmatize(token, pos=wordnet.VERB) for token in tokens]
    return lemmatized_tokens
```

```python
lemmatized_tags_tokens = [lemmatize_tokens(tokens) for tokens in filtered_tags_tokens]
```

```python
lemmatized_tags_tokens
```

```python
from gensim.models import Word2Vec
```

```python
model = Word2Vec(sentences=lemmatized_tags_tokens, vector_size=100, window=5, min_count=1, sg=1)
```

```python
def embed_tokens(tokens, model):
    # Initialize an empty array to store vectors
    vectors = []
    # Iterate over tokens in the list
    for token in tokens:
        # Check if the token is in the vocabulary of the model
        if token in model.wv.key_to_index:
            # Get the vector representation of the token
            vector = model.wv[token]
            vectors.append(vector)
    # Average the vectors to get a single vector representation for the list of tokens
    if vectors:
        return np.mean(vectors, axis=0)
    else:
        return np.zeros(model.vector_size)  # Return zero vector if no tokens are found in the model

# Create embedded vectors for each list of tokens
embedded_vectors = [embed_tokens(tokens, model) for tokens in lemmatized_tags_tokens]

```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(embedded_vectors).shape
```

```python
similarity = cosine_similarity(embedded_vectors)
```

```python
def recommend(product):
    product_index = df_shuffled[df_shuffled['title'] == product].index[0]
    distances = similarity[product_index]
    products_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:10]
    for i in products_list:
        print(df_shuffled.iloc[i[0]].title)
```

```python
recommend('Mi X Series 125')
```

```python
def recommend(title):
    top_n = 20
    product_index = new_df[new_df['title'] == title].index[0]
    distances = similarity[product_index]
    product_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:90]

    result = []
    unique_products = set()
    count = 0

    for i in product_list:
        if count >= top_n:
            break

        product_details = combined_df.iloc[i[0]].description

        # Check if the product details are not already in the set
        if product_details not in unique_products:
            product_details = combined_df.iloc[i[0]].description
            # Add the product details to the set to ensure uniqueness
            unique_products.add(product_details)
            count += 1
            temp = {
                "index" : i[0],
                "title" : new_df.iloc[i[0]].title,
            }
            result.append(temp)

           
    print(result)
```

```python
recommend('DELL Intel Core i3')
```
