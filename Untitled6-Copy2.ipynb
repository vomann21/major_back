```python
def recommend(title):
    product_index = new_df[new_df['title'] == title].index[0]
    product_type = new_df[new_df['title'] == title]['type'].iloc[0]  # get the type of the product
 # get the type of the product
    distances = similarity[product_index]
    product_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:]

    result_amazon = []
    result_flipkart = []
    unique_products = set()

    for i in product_list:
        product_details = combined_df.iloc[i[0]].description

        # Check if the product details are not already in the set
        if product_details not in unique_products and new_df.iloc[i[0]].type == product_type:
            unique_products.add(product_details)
            temp = {
                "index" : i[0],
                "title" : new_df.iloc[i[0]].title,
                "price" : combined_df.iloc[i[0]].price,
                "image" : combined_df.iloc[i[0]].image,
                "product_link": combined_df.iloc[i[0]].product_link,
                "brand" : new_df.iloc[i[0]].brand,
                "platform" : new_df.iloc[i[0]].platform,
                "description" : new_df.iloc[i[0]].description
            }

            # Add the product to the respective platform list
            if new_df.iloc[i[0]].platform.lower() == 'amazon' and len(result_amazon) < 10:
                result_amazon.append(temp)
            elif new_df.iloc[i[0]].platform.lower() == 'flipkart' and len(result_flipkart) < 20:
                result_flipkart.append(temp)

        # Break the loop if we have found 10 products for both platforms
        if len(result_amazon) >= 10 and len(result_flipkart) >= 20:
            break

    ans = {
        "recommended_products_amazon": result_amazon,
        "recommended_products_flipkart": result_flipkart
    }        
    for i in result_amazon:
        print(i['title'], i['image'])
    print('<----------------------------------------->')
    print('<----------------------------------------->')
    print('<----------------------------------------->')
    for i in result_flipkart:
        print(i['title'], i['image'])
```
